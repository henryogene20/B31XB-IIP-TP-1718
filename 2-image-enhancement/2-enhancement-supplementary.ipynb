### Q.2 Contrast streching 

** Define your function which stretch the image contrast, `cont_str`, using minimum and maximum values from the image **

**This function should accept the input image, calculates the minimum and maximum and rescales the intensities.**

** Return the image as uint8 **

```python
def cont_str (image):
    minimum = np.amin(image)
    maximum = np.amax(image)
    height, width, depth = image.shape
    for i in range(0, height):
        for j in range(0, width):
            image[i,j] = (image[i,j] - minimum)/(maximum - minimum)*255  
    return image.astype('uint8')
```

### Q.3 Histogram equalization
Histogram equalization can be used to enhance the contrast in images.

The idea behind the histogram equalization process is that each gray level should have an equal contribution in the image. The function allowing such transformation is in fact the cumulative density function.

**Define your own histgram equalization function, `hist_eq`**

**This function should accpet the image, max_int and min_int and return the transformed image and its cdf**

```python
def hist_eq (image, min_int, max_int):
    hist, bins = np.histogram(image.flatten(), 256, [0,255])
    hist = hist/sum(hist)
    cum_sum = np.cumsum(hist)
    tf = min_int + (max_int - min_int) * cum_sum
    trans_image = np.zeros_like(image)
    for i in range(0, 691):
        for j in range(0, 1048):
            trans_image[i,j] = tf[image[i,j]]
    trans_hist = exposure.histogram(trans_image)
    trans_hist = trans_hist/sum(trans_hist)
    trans_cum_sum = np.cumsum(trans_hist)
    return trans_image, trans_cum_sum
```

** Explain the process of histogram matching in your own words**

Histogram matching is a general case of histogram equalization (or, to put it the other way, histogram equalization is the specific case of the histogram matching). It is a type of image transformation that transforms an image in such a way so that the obtained image's histogram matches the specific one. The difference is in the transfer function: The histogram matching method uses the transfer function that considers the ratio of the input and the output images' probability density functions (whereas in the histogram equalization case the transfer function was a constant, which makes sense, because we want the output image's histogram to be as "flat" as possible). Practically, it can be done as the cascade transformation that consists of histogram equalization followed by the inverse histogram equalization, in order to obtain the desired histogram.


```python
def find_inverse_mapping(mapping):
    # Initialise the inverse
    inv_mapping = np.zeros(mapping.shape)
    # Find the index that we can affect for sure
    idx_inv_map = np.unique(mapping)
    
    # For each index
    for idx in idx_inv_map:
        # Find the corresponding index value
        int_val = np.where(mapping == idx)[0][0]
        # Affect this value for the inverse mapping
        inv_mapping[idx] = int_val
        
    # Find the first index which is not zero
    first_elt_idx = np.nonzero(inv_mapping)[0][0]
    for i, elt in enumerate(inv_mapping[first_elt_idx:-1]):
        if (elt == 0):
            inv_mapping[i + first_elt_idx] = inv_mapping[i + first_elt_idx - 1]
        
    return inv_mapping.astype('uint8')

```

** Explain the difference between log transform enhancement and gamma transform **


Log transformation is used in order to expand the range of dark pixels in an image and to shrink the range of the bright pixels (as opposed to inverse log function, which does the opposite). Gamma transformation uses the exponential transfer function, but we can obtain a bigger variety of transformation curves by inserting different gamma values in the formula. Using the gamma factor less than one, the resulting image is brighter, whereas for the gamma factor greater than one, the output image is darker. This is only the result of suppresing one range and expanding the other, and vice versa.

** Explain how sigmoid function enhances the image **

Transformation using the sigmoid function is another example of non-linear image enhancement. In the sigmoid function, we input the two parameters that specify the width of the input intensity range and the intensity around which that range is centered. By varying these two, we can obtain different sigmoid functions, and therefore, different output images. The idea is to specify the range of intensities in the input image that should be expanded in the output, while the intensities outside the range are suppressed. That way we can enhance the output image by improving its contrast.
